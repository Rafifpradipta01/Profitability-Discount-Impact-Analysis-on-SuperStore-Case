```python
# Loop untuk memeriksa setiap kolom yang bertipe data 'object'
for col in df.select_dtypes(include=['object']).columns:
    print(f"=== Kolom: {col} ===")
    print(f"Jumlah Nilai Unik: {df[col].nunique()}")
    
    # Menampilkan 5 nilai unik pertama sebagai sampel pemeriksaan
    print("Sampel Nilai Unik:")
    print(df[col].unique()[:5])
    print("-" * 50)
```

```python
# 1. Ubah tipe data ke datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# 2. Cek apakah ada data yang Ship Date-nya LEBIH KECIL dari Order Date
inkonsisten_tanggal = df[df['Ship Date'] < df['Order Date']]
print(f"Jumlah data tanggal inkonsisten: {len(inkonsisten_tanggal)}")
```

```python
df.dtypes
```

```python
# Cek ringkasan statistik (perhatikan nilai 'min' untuk Sales dan Quantity)
df[['Sales', 'Quantity', 'Profit']].describe()

# Atau filter langsung untuk melihat data yang tidak masuk akal
inkonsisten_sales = df[df['Sales'] <= 0]
inkonsisten_qty = df[df['Quantity'] <= 0]

print("=== HASIL CEK INKONSISTEN SALES ===")
print (inkonsisten_sales)
print("\n" + "="*50 + "\n")
print("=== HASIL CEK INKONSISTEN QTY ===")
print (inkonsisten_qty)
```

```python
# Menghitung batas atas untuk Outlier Sales
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
batas_atas = Q3 + 1.5 * IQR

# Filter untuk melihat data yang nilainya di luar kewajaran
outlier_sales = df[df['Sales'] > batas_atas]

print(f"Batas atas penjualan normal: {batas_atas}")
print(f"Jumlah data outlier (sangat tinggi): {len(outlier_sales)}")
print("\n" + "="*50 + "\n")
print("Sampel data outlier:")
outlier_sales[['Customer Name', 'Product Name', 'Sales', 'Quantity', 'Profit']].head()
```

```python
# Menghitung batas atas untuk Outlier Quantity
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1
batas_atas = Q3 + 1.5 * IQR

# Filter untuk melihat data yang nilainya di luar kewajaran
outlier_quantity = df[df['Quantity'] > batas_atas]

print(f"Batas atas penjualan normal: {batas_atas}")
print(f"Jumlah data outlier (sangat tinggi): {len(outlier_sales)}")
print("\n" + "="*50 + "\n")
print("Sampel data outlier:")
outlier_quantity[['Customer Name', 'Product Name', 'Sales', 'Quantity', 'Profit']].head()
```

```python
# Menghitung batas atas untuk Outlier Profit
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
batas_atas = Q3 + 1.5 * IQR

# Filter untuk melihat data yang nilainya di luar kewajaran
outlier_profit = df[df['Profit'] > batas_atas]

print(f"Batas atas penjualan normal: {batas_atas}")
print(f"Jumlah data outlier (sangat tinggi): {len(outlier_sales)}")
print("\n" + "="*50 + "\n")
print("Sampel data outlier:")
outlier_profit[['Customer Name', 'Product Name', 'Sales', 'Quantity', 'Profit']].head()
```

```python
# Cek rata-rata profit berdasarkan besaran diskon
analisis_diskon = df.groupby('Discount')['Profit'].mean().reset_index()

print("=== RATA-RATA PROFIT BERDASARKAN DISKON ===")
print(analisis_diskon)
```
